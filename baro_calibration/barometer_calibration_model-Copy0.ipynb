```python
%pylab inline --no-import-all
from scipy import linspace, polyval, polyfit, stats, randn
import matplotlib.pyplot as plt
import numpy as npy

def getBaseTemp(dataset1, dataset2):
    for x in range(len(dataset1)):
        if(dataset1[x] >=20):
            return dataset2[x]
    return dataset2[len(dataset1) - 1]

def readData(filename, headerLen, cols):
    return npy.genfromtxt("./data/" + filename,skip_header=headerLen,delimiter=",", usecols=(cols), unpack = 1)

CORRECTION_POLY_DEG = 3

[temp1, pressure1, altitude1]  = readData("erictest_1.csv",headerLen=4, cols=[7,6,5])

#[temp3, pressure3, altitude3]  = readData("Eric_freezer.csv",headerLen=2, cols=[7,6,5])

[x1, y1] = [temp1, pressure1] 

y1 -= getBaseTemp(x1 , y1)
p1 = polyfit(x1, y1, CORRECTION_POLY_DEG)
p = p1

print p
print "--------------------"
 
            

print '<object id=\"0x9C07C0AC\" name=\"RevoSettings\">'
print '<field name=\"BaroTempCorrectionPolynomial\" values=\"{a},{b},{c},{d}\"/>'.format(a= -p[3],b=-p[2],c=-p[1],d=-p[0])
print '</object>'


m = "s"
m2 = "^"
s = 2

datasets =  [('#AA0000', '#AA9999', x1, y1, altitude1, "testset")]

refAlt = 44330.0 * (1.0 - pow((99.59700775) / 101.3250, (1.0 / 5.255)))

for c1, c2, xx, yy, alt, l1 in datasets:
    leg = []
    fig, (ax, ax2, ax3) = plt.subplots(figsize=(20,12), nrows=3)
    #ax.scatter(xx,yy,s=s, c=c1, marker=m,alpha=0.4,linewidth=0, )
    
    #Plot original data
    ax.plot(xx,yy, alpha = .1, c=c2, linewidth=1)
    ax3.plot(xx,alt, c=c2, linewidth=1)
    leg.append(l1)
    ycorrectionFactor = polyval(p,xx)
    #thFittedcurveCoeffs = polyfit(xx,yy,CORRECTION_POLY_DEG)
    #thFittedCurveValues = polyval(thFittedcurveCoeffs,xx)
    ycorrected = yy - ycorrectionFactor
    #plot corrected value
    #ax.scatter(xx,0 - ycorrected, s=s, c=c2, marker=m2,linewidth=1)
    ax.plot(xx,ycorrected, c=c2, linewidth=1)
    #ax.plot(xx,ycorrectionFactor, c=c2, linewidth=3)
    leg.append(l1 + "/CAL")
    #ax.plot(xx,pty, c=c2, linewidth=1)
    #leg.append(l1 + "/TH")
    
    #set a common x range
    ax.set_xlim(-10, +40)
    ax2.set_xlim(-10, +40)
    
    # calculate an hypotetical altitude error considering a 99.59700775 base pressure
    alt = (refAlt - 44330.0 * (1.0 - pow((ycorrected+99.59700775) / 101.3250, (1.0 / 5.255)))) 
    ax2.plot(xx,alt, c='b', linewidth=1)
    
    #ax.set_ylim(-0.05, 0.05)
    #ax2.set_ylim(-1, 1)
    
    plt.grid(b=True, which='both',color='0.5',linestyle='-')
    ax.legend(leg)
    ax2.legend(["est Alt. error"])
```
