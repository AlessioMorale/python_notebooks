```python
%pylab inline --no-import-all
import pandas as PD
from scipy import linspace, polyval, polyfit, stats, randn
from scipy.signal import butter, lfilter
import matplotlib.pyplot as plt
import numpy as npy

import numpy

def getBaseTemp(dataset1, dataset2):
    for x in range(len(dataset1)):
        if(dataset1[x] >=20):
            return dataset2[x]
    return dataset2[len(dataset1) - 1]

def readData(filename, headerLen, cols):
    return npy.genfromtxt("./gyro_data/" + filename,skip_header=headerLen,delimiter=",", usecols=(cols), unpack = 1)

CORRECTION_POLY_DEG = 1

[tempg1, gx1, gy1, gz1]  = readData("uncal3_Gyro_2013-12-24_19-19-08.csv",headerLen=2, cols=[5,6,7,8])
[tempa1, ax1, ay1, az1]  = readData("uncal3_Accel_2013-12-24_19-19-08.csv",headerLen=2, cols=[5,6,7,8])
[tempb1, pre1, alt3]  = readData("uncal3_Barometer_2013-12-24_19-22-32.csv",headerLen=2, cols=[7,6,5])



#[tempg1, gx1, gy1, gz1]  = readData("Gyro_2013-12-23_14-45-25_calib.csv",headerLen=2, cols=[5,6,7,8])
#[tempa1, ax1, ay1, az1]  = readData("Accel_2013-12-23_14-45-25_calib.csv",headerLen=2, cols=[5,6,7,8])
#[tempb1, pre1, alt3]  = readData("Barometer_2013-12-23_14-46-00_calib.csv",headerLen=2, cols=[7,6,5])


m = "s"
#m2 = "^"
m2 = "s"
s = 2
freq = 2
b, a = butter(4, 0.1, btype = 'low')


sets =  [
         ('#AA0000', '#AA9999', tempg1, gx1, "gx1"),
         ('#00AA00', '#66AA66', tempg1, gy1, "gy1"),
         ('#00AAAA', '#99AAAA', tempg1, gz1, "gz1"),
         ('#AAAA00', '#AAAA66', tempa1, ax1, "ax1"),
         ('#EE0000', '#EE9999', tempa1, ay1, "ay1"),
         ('#EE00EE', '#EE99EE', tempa1, az1, "az1"),
         ]


    
```

```python
print(__doc__)

# Author: Mathieu Blondel
# License: BSD 3 clause

import numpy as np
import pylab as pl
from scipy import sqrt
from sklearn.linear_model import Ridge

degree = 1
pbiases = {}

for [c1, c2, x, iy, label] in sets:
    fig, (ax, ax2) = plt.subplots(figsize=(20,10), nrows=2)
    y = iy
    y2 = lfilter(b, a, iy)  # standard filter
    
    pbias = polyfit(x ,y , degree)
    predicty = polyval(pbias,x)
    
    pbiases[label] = pbias
    
    pbias2 = polyfit(x ,y , degree + 1)
    predicty2 = polyval(pbias2,x)
    
    ax.plot(x, predicty)
    ax.plot(x, predicty2)
    ax.scatter(x, y,s=s, c=c1, marker=m,alpha=1,linewidth=0 )
    ax.legend(["degree %d" % degree,"degree %d" %  (degree + 1),label])
    rebiasedY = y2-predicty;
    rebiasedY2 = y2-predicty2;
    ax2.scatter(x, y-predicty,s=s, c=c1, marker=m,alpha=1,linewidth=0 )
    ax2.plot(x,rebiasedY)
    #ax2.plot(x,rebiasedY2)
    ax2.legend([label + "/cor %d" % degree, label + "/cor %d" % (degree+1),label +"rebiased"])
    
    
    ax.grid(b=True, which='both',color='0.5',linestyle='-')
    ax2.set_ylim(-0.10, 0.10)
    ax2.grid(b=True, which='both',color='0.5',linestyle='-')
    print label, pbias

```
