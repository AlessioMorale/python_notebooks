```python
%pylab inline --no-import-all
from scipy import linspace, polyval, polyfit, stats, randn
import matplotlib.pyplot as plt
import numpy as numpy

def getBaseTemp(dataset1, dataset2):
    for x in range(len(dataset1)):
        if(dataset1[x] >=20):
            return dataset2[x]
    return dataset2[len(dataset1) - 1]

def readData(filename, headerLen, cols):
    return numpy.genfromtxt("./data/" + filename,skip_header=headerLen,delimiter=",", usecols=(cols), unpack = 1)

CORRECTION_POLY_DEG = 3

[temp1, pressure1, altitude1]  = readData("Barometer_2013-12-09_19-30-21.csv",headerLen=2, cols=[9,6,5])
[temp2, pressure2, altitude2]  = readData("Barometer_2013-12-10_01-03-37.csv",headerLen=2, cols=[7,6,5])

[x1, y1, p1] = [temp1, altitude1,numpy.ndarray(CORRECTION_POLY_DEG+1)] 
[x2, y2, p2] = [temp2, altitude2,numpy.ndarray(CORRECTION_POLY_DEG+1)] 

for tup in [[x1,y1,p1],[x2,y2,p2]]:
    tup[1] -= getBaseTemp(tup[0] , tup[1])

m = "s"
m2 = "^"
s = 2




sets =  [
         ('#AA0000', '#AA9999', x1, y1, "AM_20-cal"),
         ('#00AA00', '#66AA66', x2, y2, "AM_20-cal+bias"),
         ]
refAlt = 44330.0 * (1.0 - pow((102.7800064) / 101.3250, (1.0 / 5.255)))

for c1, c2, xx, yy, l1 in sets:
    leg = []
    fig, ax = plt.subplots(figsize=(20,10), nrows=1)
    ax.scatter(xx,yy,s=s, c=c1, marker=m,alpha=0.4,linewidth=0, )
    #ax.plot(xx,yy, alpha = .5, c=c2, linewidth=1)
    leg.append(l1)
    ax.set_xlim(-10, +40)
    ax.set_ylim(-5, 5)
    plt.grid(b=True, which='both',color='0.5',linestyle='-')
    ax.legend(leg)
```
